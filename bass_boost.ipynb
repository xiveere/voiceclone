```python
from pydub import AudioSegment
from os import listdir
import numpy as np
import math

song_dir = "generated_audio"
attenuate_db = 0 # Reduce volume of rest of track
accentuate_db = 2 # Boost volume of bass


def bass_line_freq(track):
    sample_track = list(track)

    # c-value
    est_mean = np.mean(sample_track)

    # a-value
    est_std = 3 * np.std(sample_track) / (math.sqrt(2))

    bass_factor = int(round((est_std - est_mean) * 0.01))

    return bass_factor

wavs = [file for file in listdir(song_dir) if 'wav' in file]

for filename in  wavs:
    sample = AudioSegment.from_wav(song_dir + "/" + filename)
    filtered = sample.low_pass_filter(bass_line_freq(sample.get_array_of_samples()))

    combined = (sample - attenuate_db).overlay(filtered + accentuate_db)
    combined.export("generated_audio/bass_boosted/" + filename.replace(".wav", "") + "-bass_boost.wav", format="wav")
```
